```python
# Create a CountVectorizer for parsing/counting words
count_vectorizer = CountVectorizer(stop_words=stopwords, min_df = 2, max_df=0.7)

count_vectorizer.fit(df.comments)
```

```python
# Create the term-document matrix
# Transpose it so the terms are the rows
doc_word = count_vectorizer.transform(df.comments).transpose()
```

```python
doc_word.shape
```

```python
# Convert sparse matrix of counts to a gensim corpus
corpus = matutils.Sparse2Corpus(doc_word)
id2word = dict((v, k) for k, v in count_vectorizer.vocabulary_.items())
len(id2word)
```

# model with 3 topics 

```python
# Create lda model (equivalent to "fit" in sklearn)
lda = models.LdaModel(corpus=corpus, num_topics=3, id2word=id2word)
```

```python
# save model
lda.save('lda_3topics.model')
```

```python
lda.print_topics()
```

2 distinct topics: Rooms and Location

```python
# using sklearn

# using tfidf vectorizer 
tf = TfidfVectorizer(stop_words=stopwords, min_df = 2, max_df=0.7)

word_matrix = tf.fit_transform(df['comments'])

vocab = tf.get_feature_names()
```

# Model with 10 topics

```python
# 10 topics
lda_tf = LatentDirichletAllocation(n_components=10, random_state=0)
lda_tf.fit(word_matrix)
```

```python
def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
```

```python
vocab = tf.get_feature_names()
display_topics(lda_tf, vocab, 10)
```

A lot of repeating words like clean and apartment. Adding apartment to stopwords.

```python
lda_tf = LatentDirichletAllocation(n_components=6, random_state=0)
lda_tf.fit(word_matrix)
```

```python
vocab = tf.get_feature_names()
display_topics(lda_tf, vocab, 10)
```

Topic 1: Booking Process
Topic 2: Location
Topic 3: Host communications
