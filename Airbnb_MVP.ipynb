# Airbnb Reviews Topic Modeling/Recommendation System

The goal of this project is to do Topic Modeling on Airbnb reviews and create a recommendation system based on the reviews. I used Tf-idf with NMF for topic modeling with 3 topics. I am also going to create a recommendation system using Streamlit so users can find similar listings based on a previous review. 

# Word clouds for topics 1, 2, and 3

Topic 1 seems to be about the location of the Airbnb(being close to public transportation) and attractions.
Topic 2 seems to be about the interior of the Airbnb.
Topic 3 seems to be about the booking logistics.
